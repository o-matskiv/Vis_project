```python
import pandas as pd
import numpy as np
import altair as alt
```

# Код знизу - приклад того,як я буду працювати з кожним файлом (по роках). А опісля цього я всю зібрану інформацію зберу разом і буду порівнювати загальні тенденції

```python

df = pd.read_excel (r'2014 Worker Fatalities.xlsx')
df.head()
```

# Нещасні випадки за статтю

```python
df_sex = df.copy(deep=False)
df_sex = df_sex.groupby('Gender').size()
df_sex = df_sex.reset_index()
df_sex
```

# Нещасні випадки за штатом

```python
df_sex = df.copy(deep=False)
df_sex = df_sex.groupby('event_state').size()
df_sex = df_sex.reset_index()
df_sex.head()
```

# Нещасні випадки за віком

```python
df_age1 = df.copy(deep=False)
df_age1 = df_age1.groupby('age').size()
df_age1 = df_age1.reset_index()
df_age1
```

# Завелика диференціація за віком, доречно розбити на групи


```python
df_age2 = df.copy(deep=False)

df_age2 = df_age2.groupby(pd.cut(df["age"], np.arange(0, 90, 5))).size()


df_age2
```

```python
df_test_state = df.copy(deep=False)


df_test_state = (df_test_state.groupby(['event_state', 'sex']).size()).to_frame().reset_index()
df_test_state = df_test_state.set_axis([*df_test_state.columns[:-1], 'total'], axis=1, inplace=False)
df_test_state.head()
```

# Нещасні випадки за статтю за 2019 рік

```python
chart1 = alt.Chart(df_test_state,title='Male/female deathes per state').mark_bar().encode(
    x = alt.X('event_state:N'),
    y = alt.Y('total:Q'),
    color = alt.Color('sex:N'),
    tooltip = alt.Tooltip('total')
)
chart1

```
