```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_excel (r'Year_Sex.xlsx')
df.head()
```

# Загальна статистика смертей за гендером за всі роки

```python
df_m = df.melt(id_vars=['Year'], var_name='Sex', value_name='Total')
df_m.head()
```

```python
chart1 = alt.Chart(df_m,title='Male/female deathes').mark_bar().encode(
    x = alt.X('Sex:N'),
    y = alt.Y('Total:Q'),
    color = alt.Color('Sex:N'),
    tooltip = alt.Tooltip('Total')
).properties(
    width=200,
    height=200)
chart1

```

```python
chart2 = alt.Chart(df_m,title='Male/female deathes per year').mark_bar().encode(
    x = alt.X('Year:N'),
    y = alt.Y('Total:Q'),
    color = alt.Color('Sex:N'),
    tooltip = alt.Tooltip('Total')
).properties(
    width=400,
    height=200)
chart2


```
